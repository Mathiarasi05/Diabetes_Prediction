```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
import joblib
import warnings
warnings.filterwarnings("ignore")

sns.set_style("whitegrid")
```

```python
fig, axes = plt.subplots(3, 3, figsize=(15, 12))
features = diabetes_dataset.columns[:-1]

for idx, feature in enumerate(features):
    if idx >= 9: break  # Prevents error if there are more than 9 features
    
    row, col = idx // 3, idx % 3
    
    for outcome in [0, 1]:
        # Filter data for the specific outcome and feature
        data = diabetes_dataset[diabetes_dataset['Outcome'] == outcome][feature]
        
        # Plotting the histogram
        axes[row, col].hist(data, alpha=0.6, bins=20, 
                            label=f'Outcome {outcome}',
                            color=['#2ecc71', '#e74c3c'][outcome])
        
    # Styling and Labels
    axes[row, col].set_title(feature, fontweight='bold')
    axes[row, col].set_xlabel('Value')
    axes[row, col].set_ylabel('Frequency')
    axes[row, col].legend()

    # Remove unused subplots
for i in range(len(features), 9):
    fig.delaxes(axes.flatten()[i])

plt.tight_layout() # Prevents labels from overlapping
plt.show()
```

```python
X = diabetes_dataset.drop(columns='Outcome', axis=1)
Y = diabetes_dataset['Outcome']
print(X)
print(Y)
```

```python
scaler=StandardScaler()
X = scaler.fit_transform(X)
X
```

```python

# Splitting the data into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)

# Printing the shapes to verify the split
print(X.shape, X_train.shape, X_test.shape)
```

```python
# Initializing the SVM classifier with a linear kernel
classifier = svm.SVC(kernel='linear', probability=True)

# Training the classifier
classifier.fit(X_train, Y_train)
```

```python
# Training the model and predicting on training data
X_train_prediction = classifier.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
print('Accuracy score on training data : ', training_data_accuracy)
```

```python
# Predicting on test data and calculating test accuracy
X_test_prediction = classifier.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
print('Accuracy score on test data : ', test_data_accuracy)
```

```python
# Calculating detailed performance metrics
precision = precision_score(Y_test, X_test_prediction)
recall = recall_score(Y_test, X_test_prediction)
f1 = f1_score(Y_test, X_test_prediction)
print(precision, recall, f1)
```

```python
# Generate the confusion matrix data
# Note: Ensure y_test and x_test_prediction are already defined in your environment
cm_svm = confusion_matrix(Y_test, X_test_prediction)
cm_svm
```

```python
# Plotting the Confusion Matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm_svm, annot=True, fmt='d', cmap='Blues', square=True,
            xticklabels=['Non-Diabetic', 'Diabetic'],
            yticklabels=['Non-Diabetic', 'Diabetic'])

# Adding Labels and Title
plt.title('Confusion Matrix - SVM', fontsize=14, fontweight='bold')
plt.ylabel('Actual')
plt.xlabel('Predicted')

# Display the plot
plt.show()
```

```python
# Initialize the classifier with 100 trees and a fixed random state for reproducibility
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=2)

# Train (fit) the model using the training data
rf_classifier.fit(X_train, Y_train)
```

```python
# Generate predictions on the training data
X_train_prediction_rf = rf_classifier.predict(X_train)
```

```python
# Calculate accuracy on the training set
training_data_accuracy_rf = accuracy_score(X_train_prediction_rf, Y_train)
print(f"Training Data Accuracy: {training_data_accuracy_rf}")
```

```python
# Generate predictions on the unseen test data
X_test_prediction_rf = rf_classifier.predict(X_test)

# Calculate accuracy on the test set
test_data_accuracy_rf = accuracy_score(X_test_prediction_rf, Y_test)
print(f"Test Data Accuracy: {test_data_accuracy_rf}")
```

```python
# Calculate the precision score
precision_rf = precision_score(Y_test, X_test_prediction_rf)
```

```python
# Assuming y_test are true labels and x_test_prediction_rf are your model's predictions
precision_rf = precision_score(Y_test, X_test_prediction_rf)
recall_rf = recall_rf = recall_score(Y_test, X_test_prediction_rf)
f1_rf = f1_score(Y_test, X_test_prediction_rf)

# Print the metrics clearly
print(f"Precision: {precision_rf:.4f}")
print(f"Recall:    {recall_rf:.4f}")
print(f"F1 Score:  {f1_rf:.4f}")
```

```python
# Generate the matrix
cm_rf = confusion_matrix(Y_test, X_test_prediction_rf)
print(cm_rf)
```
